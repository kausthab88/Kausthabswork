### 3.1 Do you think there is evidence that mean moisture contents in both types of shingles are within the permissible limits? State your conclusions clearly showing all steps.

* In order to solve this part of the problem, we solve separately for A & B. We do a 1 Sample T-Test for and A & B both. 

The $H_0$ <= 0.35 and 
    $H_A$> 0.35 and $\alpha$ = 0.05
    
 On solving for A, we see that the p-value (or p_value_a) is 0.15 which is greater than $\alpha$. Thus, we fail to reject $H_0$. So for Shingles A, we conclude with a confidence level of 95%, the moisture content is within the permissible limit of 0.35 pound per 100 square feet
 
 On performing the T-Test for Shingles B, the p-value (or p_value_b) is 0.00418 which is less than $\alpha$ and so we  reject $H_0$. So for Shingles B, we conclude with a confidence level of 95%, the moisture content lies outside (or greater than) the permissible limit of 0.35 pound per 100 square feet

### 3.2 Do you think that the population means for shingles A and B are equal? Form the hypothesis and conduct the test of the hypothesis. What assumption do you need to check before the test for equality of means is performed?

 * To answer the first part of the question, we made the following assumptions:
             
     $H_0$: $\mu_A$ = $\mu_B$  (Population mean of Shingles A and Shingles B are equal)                               

     $H_A$: $\mu_A$ $\neq$ $\mu_B$
     
     and $\alpha$ = 0.05
         
     Since, we ae comparing both Shingles A and Shingles B, it will be a two-tailed test for independent samples. After performing the test, we see that the p-value is 0.201 which is more than $\alpha$. Thus we fail to reject $H_0$. 
      
      
    Now, some assumptions that we need to check for before testing the equaliy are:

    1. The population standard deviations are assumed to be equal since the sample standard deviations tend to 1
    
    2. We also assume that the data is continuous
    
    3. The data follow normal probability distribution
    
    4. The two samples are independent. There is no relationship between the individuals in one sample as compared to the other
    
    5. Both samples are simple random samples from their respective populations. Each individual in the population has an equal probability of being selected in the sample.

$H_0$: $\mu_A$ = $\mu_B$
